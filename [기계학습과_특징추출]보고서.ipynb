심성빈(201904213), 김찬민(201904201),한지은(202004245),성윤주(202184028)

1. 개요

분석의 목적은 경제적인 소득과 공연예술활동 소비간의 관계를 파악하는 것입니다.
시 도 단위 인구수, 연소득 순위와 각종 공연 매출액을 비교 분석하여 경제적 여건이 문화 소비에 어떤 영향을 미치는지 탐색하고 고소득 지역이 높은 문화 소비를 보이는지 등 경제적 여건이 다양한 문화 소비로 이어지는지 파악하여 최종적으로 지역별 경제 상황과 공연예술활동 소비의 관계를 시각화 하는 것이 목표입니다.

2. 데이터 수집

데이터 분석 프로젝트를 수행하기 위한 수집과정에서 신뢰성을 고려하여공공 데이터를 전략적으로 활용 할 것으로 결정하였습니다. 이를 위해 관련 분야에 대한 공공 데이터 소스를 식별하고, 데이터의 적합성 평가 과정을 선행하여 최종적으로 분석에 사용될 데이터 세트를 선정, 혹은 필요한 정보를 수집하여 표준화 과정을 거친 데이터 셋을 이용할 계획입니다.

데이터셋의 기반이 되는 공공 데이터 정보는 행정안전부와 한국문화정보원의 연령별 인구 현황, 국내 공공 도서관, 박물관, 미술관, 문화 회관 그리고 KOPIS 공연시설 통계 데이터, 문화 빅데이터 플랫폼의 거주자 소득대비 문화시설 지도 데이터의 전국 시도단위 인구수, 연소득 순위, 각 공연예술별 매출액 등 여러 기관의 데이터가 사용됩니다.

사용될 데이터는 csv형태로 각각의 정보마다 여러 다른 항목을 담고있습니다. 이러한 여러 항목들을 정제하여 최종적으로 대한민국의 각 지역(시도 및 시군구) 단위로 인구수와 연소득, 문화시설(오페라, 뮤지컬, 클래식 공연 등)의 수량 및 비율, 그리고 각 공연예술별 매출액의 특성을 가진 데이터셋으로 만들어 분석프로젝트에 이용할 것입니다.
